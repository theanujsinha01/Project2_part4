PROBLEM OBJECTIVE: RESTAURANT CHAINS

(1)IDENTIFY IF THERE ARE ANY RESTAURANT CHAINS PRESENT IN THE DATASET.

(2)ANALYZE THE RATINGS AND POPULARITY OF DIFFERENT RESTAURANT CHAINS.


```python
#IDENTIFY IF THERE ARE ANY RESTAURANT CHAINS PRESENT IN THE DATASET.
resta = data['Restaurant Name'].value_counts()
resta_chain = resta[resta>1]



resta_chain
```

```python
# #ANALYZE THE RATINGS AND POPULARITY OF DIFFERENT RESTAURANT CHAINS.
chain_rating = data[data['Restaurant Name'].isin(resta_chain.index)].groupby('Restaurant Name')

# Calculate the average rating for each restaurant chain
average_rating = chain_rating['Aggregate rating'].mean()

chain_analysis = pd.DataFrame({
    'Average Rating': average_rating
   
})

# Sort the DataFrame by 'Average Rating'
chain_analysis = chain_analysis.sort_values(by=['Average Rating'], ascending=[False])

# Display the results
chain_analysis.head(10)
```
